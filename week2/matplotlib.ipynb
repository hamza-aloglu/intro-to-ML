```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.arange(0, 10, 0.1)
sinValues = np.sin(data)
plt.plot(data, sinValues) # (x, y)

plt.title("sin graph")
plt.show()
```

```python
xData = np.random.rand(50)
yData = np.random.rand(50)

colors = np.random.rand(50)
sizes = np.random.randint(50, 100, 50)

plt.scatter(xData, yData, c=colors, s=sizes)

plt.title("point chart")

plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)

plt.title("histogram chart")
plt.show()
```

```python
x = ['x0', 'x1', 'x2', 'x3']
y = np.random.randint(0, 10, 4)
plt.bar(x, y)
plt.title('Column chart')
plt.show()
```

```python
data= ['sector1', 'sector2', 'sector3', 'sector4', 'sector5']
values= np.random.randint(0, 100, 5)
plt.pie(values, labels=data)
plt.title('pie chart')
plt.show()
```
